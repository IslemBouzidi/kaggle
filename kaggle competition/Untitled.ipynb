```python
num_features = len(features[1:])
num_plots_per_row = 3
num_rows = num_features // num_plots_per_row + (num_features % num_plots_per_row > 0)

fig, axs = plt.subplots(num_rows, num_plots_per_row, figsize=(15, 4*num_rows))
fig.subplots_adjust(hspace=0.5)  # Adjust the vertical spacing between subplots

for i, feature in enumerate(features[1:]):
    row = i // num_plots_per_row
    col = i % num_plots_per_row

    ax = axs[row, col]
    sns.scatterplot(x=feature, y="SalePrice", data=train, marker='o', ax=ax)
    ax.set_title(f"Scatter plot with regression line for {feature}")
    ax.set_xlabel(feature)
    ax.set_ylabel("SalePrice")

# Remove empty subplots if the number of features is not a multiple of num_plots_per_row
if num_features % num_plots_per_row != 0:
    for i in range(num_features, num_rows * num_plots_per_row):
        axs[-1, i % num_plots_per_row].axis('off')

plt.tight_layout()
plt.show()
```
